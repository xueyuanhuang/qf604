```python
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
#"QCOM" stands for Qualcomm Incorporated
#"INTC" represents Intel Corporation
#"MU" is the ticker for Micron Technology, Inc.
ticker = "QCOM"
stock_data = yf.download(ticker, "1992-01-01", "2022-12-24")
qcom_price = stock_data['Adj Close']

ticker = "INTC"
stock_data = yf.download(ticker, "1992-01-01", "2022-12-24")
intc_price = stock_data['Adj Close']

ticker = "MU"
stock_data = yf.download(ticker, "1992-01-01", "2022-12-24")
mu_price = stock_data['Adj Close']
```

```python
#"IRX" is the ticker symbol for the 13-week Treasury Bill (T-bill).
ticker = '^IRX'
rate_data = yf.download(ticker, '1992-01-01', '2022-12-24')
tbill_rate = rate_data['Adj Close']
daily_tb = tbill_rate/365
```

```python
#"^GSPC" is the ticker symbol for the S&P 500 Index
ticker = '^GSPC'
start = '1992-01-01' 
end = '2022-12-24'
index_data = yf.download(ticker, start, end)
spx = index_data['Adj Close']
```

```python
qcom_return = np.log(qcom_price/qcom_price.shift(1))
intc_return = np.log(intc_price/intc_price.shift(1))
mu_return = np.log(mu_price/mu_price.shift(1))

spx_return = np.log(spx/spx.shift(1))

ex_qcom_return = qcom_return[1 : ] - daily_tb[1 : ]
ex_intc_return = intc_return[1 : ] - daily_tb[1 : ]
ex_mu_return = mu_return[1 : ] - daily_tb[1 : ]

ex_spx_return = spx_return[1 : ] - daily_tb[1 : ]
```

```python
len(qcom_return), len(spx_return)
```

```python
from statsmodels.api import add_constant
from statsmodels.regression.rolling import RollingOLS

ex_spx_return = add_constant(ex_spx_return, prepend = False) #means constant put at the end

mod1 = RollingOLS(ex_qcom_return, ex_spx_return, window = 60, expanding = False)
rolling_res1 = mod1.fit()

mod2 = RollingOLS(ex_intc_return, ex_spx_return, window = 60, expanding = False)
rolling_res2 = mod2.fit()

mod3 = RollingOLS(ex_mu_return, ex_spx_return, window = 60, expanding = False)
rolling_res3 = mod3.fit()
```

```python
rolling_res1.params.columns = ['exc_mkt_ret', 'constant']
rolling_res2.params.columns = ['exc_mkt_ret', 'constant']
rolling_res3.params.columns = ['exc_mkt_ret', 'constant']
```

```python
pd.set_option('display.max_rows', None)
rolling_res1.params.iloc[59:100, ] #### previous dates all NaN as being part of first window of length 60
```

```python
M1 = rolling_res1.params['exc_mkt_ret']
M1 = M1.iloc[59 : , ]

M2 = rolling_res2.params['exc_mkt_ret']
M2 = M2.iloc[59 : , ]

M3 = rolling_res3.params['exc_mkt_ret']
M3 = M3.iloc[59 : , ]
```

```python
import matplotlib.pyplot as plt

plt.plot(M1, linestyle = '-', linewidth = .5)
plt.plot(M2, linestyle = '--', linewidth = .5)
plt.plot(M3, linestyle = '-.', linewidth = .5)

plt.title('60-Day Moving Window Beta Estimates from 1992 to 2022')
plt.legend(['qcom', 'intc', 'mu'])
plt.show()
```
