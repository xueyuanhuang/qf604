```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
### Data are daily portfolio returns of stocks from SGX during 28 Oct 1997 through to 18 Oct 2002.
### The large stock portfolio returns (LSR) are simple daily ave return rates from 10 stocks viz.
###     Singtel, UOOB, DBS, OCBC, SIA, SPH, Jardine, HK Land, Great Eastern, and City Developments
### The small stock portfolio returns (SSR) are simple daily ave return rates from 10 stocks viz.
###     Econ Intl, Casa Holdings, Pertama Holdings, Meiban Group, Sunright Ltd, Armstrong Ind Corp, Penguin Boat, Freight Links
###     Express Holdings, Liang Huat Aluminium, and Tye Soon Ltd
### The market return rate is proxied by Straits Times Index return rate, STIR
```

```python
df = pd.read_csv('Large_Small_Day_of_Week.csv') ### need to ensure numbers are in numbers, not general or other formats in CSV
df=df.dropna()
print(df)
```

```python
dat1 = df[df['d1'] != 0] ### dat1 contains only Monday returns
dat1.reset_index(drop=True, inplace=True) 
### this is important to reset index to 0,1,2,... so that when concatenating - they align
print(dat1)
```

```python
mL=dat1['LSR'].mean()
mS=dat1['SSR'].mean()
vL=dat1['LSR'].var()
vS=dat1['SSR'].var() # .var is biased, i.e., divided by sample size
mL-mS
```

```python
pooledvar=((259)*vL + (259)*vS)/(259+259-2)
```

```python
t_diff=(mL-mS)/np.sqrt(pooledvar*(1/259+1/259))
print(t_diff)
```

```python
import scipy.stats
c = scipy.stats.t.sf(t_diff, 259+259-2) ## gives prob above t_diff
pval=c*2  ## p-value for two tailed test; reject null at 2% significance level
pval
```

```python
### Large stocks
from statsmodels.formula.api import ols
formula = 'LSR ~ d1 + d2 + d3 + d4 + d5 + STIR - 1'
results1 = ols(formula, df).fit()
print(results1.summary())
### Note: adding ' -1 ' removes constant or intercept
### Note that if we remove the  - 1, this statsmodel program still runs and shows regression results,
###   but warns of singular matrix -- other programs e.g. R would not return any result
```

```python
var1=np.var(results1.resid)*1299/1293  #multiply by 1299/1293 to get unbiased variance
print(var1,np.sqrt(var1))  ### regression residual error vola
```

```python
### Small stocks
from statsmodels.formula.api import ols
formula = 'SSR ~ d1 + d2 + d3 + d4 + d5 + STIR - 1'
results2 = ols(formula, df).fit()
print(results2.summary())
### Note: adding ' -1 ' removes constant or intercept
### Note that if we remove the  - 1, this statsmodel program still runs and shows regression results,
###   but warns of singular matrix -- other programs e.g. R would not return any result
```

```python
var2=np.var(results2.resid)*1299/1293  #multiply by 1299/1293 to get unbiased variance
print(var2,np.sqrt(var2))  ### regression residual error vola
```

```python
### There appears to be a positve Tuesday effect significantly diff from zero at 5% for large stocks
###   but a negative Monday effect significantly diff from zero at 1% for small stocks
### The fitted residual error vola for small stocks is also larger at 0.0207 versus those of large stocks at 0.0057. 
```

```python
import matplotlib.pyplot as plt
plt.plot(results1.resid,linestyle="-",linewidth=0.5)
plt.plot(results2.resid,linestyle="--",linewidth=0.5)
plt.title('Residuals from 28 Oct 1997 to 18 Oct 2002')
plt.legend(['Large Stocks','Small Stocks'])
plt.show()
```

# Form Stacked Regression and Assume Heteroskedasticity

```python
SR=pd.concat([df['LSR'],df['SSR']],axis=0); len(SR)
```

```python
df['zero']=0 ### one col of 1299 rows of zeros according to length of df
MR1=pd.concat([df['STIR'],df['zero']],axis=0)
MR2=pd.concat([df['zero'],df['STIR']],axis=0)
print(MR1.shape,MR2.shape)
## this concatenation is to ensure bigX matrix later is not singular
```

```python
Dx=pd.concat([df['d1'],df['d2'],df['d3'],df['d4'],df['d5']],axis=1)
Dx2=pd.concat([Dx,Dx],axis=0) ### 2598 rows x 5 cols of the dummies
print(Dx2)
```

```python
df2 = pd.concat([SR,MR1,MR2,Dx2],axis=1)
df2.columns=['SR','MR1','MR2','d1','d2','d3','d4','d5']
print(df2)
```

# Suppose first 1299 errors in the stacked regression has smaller variance var1, and next 1299 errors in the stacked regression has larger variance var2

```python
S1=np.identity(1299)*var1 ### 1299 x 1299
S2=np.identity(1299)*var2 ### 1299 x 1299
S11 = pd.DataFrame(S1)
S22 = pd.DataFrame(S2)
Z=np.zeros([1299,1299])
Z1 = pd.DataFrame(Z)
sigma1=pd.concat([S11,Z1],axis=0)
sigma2=pd.concat([Z1,S22],axis=0)
Sigma=pd.concat([sigma1,sigma2],axis=1)
print(np.diag(Sigma)) ### all other off diag elements are zeros
```

# Perform GLS

```python
## Above formulation avoids restricting LSR and SSR to have same beta on market. 
## Any beta restrictions on market will yield residual errors that will not match the \hat sigma_LSR, \hat sigma_SSR
## Assuming Sigma above is asymptotically convergent on true cov matrix of the residuals of LSR and SSR. 
## Then perform GLS based on the estimated variances - this is Feasible GLS

exog=pd.concat([Dx2,MR1,MR2],axis=1); exog.columns=['d1', 'd2', 'd3', 'd4', 'd5', 'MR1', 'MR2']
gls_model = sm.GLS(df2['SR'], exog, sigma=Sigma) ## Note sigma is cov matrix, not corr matrix
gls_results = gls_model.fit()
print(gls_results.summary())
```

```python
var3=np.var(gls_results.resid)*2598/2591  # get unbiased variance
print(var3,np.sqrt(var3))  ### regression residual error vola
```

```python
import matplotlib.pyplot as plt
plt.plot(gls_results.resid,linestyle="-",linewidth=1)
plt.title('Residuals from 31 August 1998 to 18 October 2002')
plt.show()
```

# Check for Heteroskedasticity

```python
### Test for Heteroskedasticity
from statsmodels.stats.diagnostic import het_breuschpagan
bp_test = het_breuschpagan(gls_results.resid,exog)
print ('Stacked Regression with Heteroskedasticity Adjustment LM-test chi-sq:', bp_test[0], 'LM-test p_value:', bp_test[1])
### There are only 6 degrees of freedom due to expl vars d2,d3,d4,d5 and MR1, MR2 as constructed. An intercept is used
### and d1 is not included to avoid dummy variable trap in this Breushpagan residual error square regression.
### Note: bp_test[2],bp_test[3] give F(6,2591) stats and F p-value. 2591 = 2598 less 7 expl vars incl intercept
```

Cannot reject null H_0 (all coeffs = 0 in regression of fitted residuals on X) of no heteroskedasticity at 10% significance level

```python
### White's Heteroskedasticity Test
from statsmodels.stats.diagnostic import het_white
wh_test = het_white(gls_results.resid,exog)
print ('LM-test chi-sq:', wh_test[0], 'LM-test p_value:', wh_test[1])
### There are only 16 degrees of freedom due to expl vars 
### d2,d3,d4,d5,MR1,MR2, (an intercept is used and thus d1 cannot be added to avoid dummy variable trap) 
### d2*MR1,d3*MR1,d4*MR1,d5*MR1, d2*MR2,d3*MR2,d4*MR2,d5*MR2, (d1*MR1 and d1*MR2 cannot be added, otherwise
###    (d1+d2+d3+d4+d5)*(MR1+MR2)=MR1+MR2 used, and leads to dummy variable trap)
### MR1^2, MR2^2. Hence 16 vars (excluding intercept) used in White heteros regression
### Note: wh_test[2],wh_test[3] give F(16,2581) stats and F p-value. 2581 = 2598 less 17 expl vars incl intercept
```

Reject null H_0 (all coeffs = 0 in regression of fitted residuals on X and cross-products of X) of no heteroskedasticity at 1% significance level

# Next check out autocorrelation in the fitted residual errors

```python
resid1_fit = sm.OLS(
    np.asarray(results1.resid)[1:], sm.add_constant(np.asarray(results1.resid)[:-1])
).fit()
print(resid1_fit.tvalues[1])
print(resid1_fit.pvalues[1])
rho1 = resid1_fit.params[1]
print(rho1)
```

```python
resid2_fit = sm.OLS(
    np.asarray(results2.resid)[1:], sm.add_constant(np.asarray(results2.resid)[:-1])
).fit()
print(resid2_fit.tvalues[1])
print(resid2_fit.pvalues[1])
rho2 = resid2_fit.params[1]
print(rho2)
```

```python
### If we use 10% 2-tailed significance level, then we reject null rho = 0 in the large stocks regressions,
###   but not in the Small stocks regression
### We therefore study autocorrelations in the residuals of the Large stock portfolio 
```

# Perform GLS below

```python
DxL=pd.concat([df['d1'],df['d2'],df['d3'],df['d4'],df['d5'],df['STIR']],axis=1)
from scipy.linalg import toeplitz
trix = toeplitz(range(len(results1.resid))) 
### trix is sq matrix with zero in diag, 1 in first off diag, 2 in 2nd off diag, etc.
sigma_gls = rho1 ** trix ### this is cov matrix of residuals with scalar constant not necessarily specified 
gls_model1 = sm.GLS(df['LSR'], DxL, sigma=sigma_gls)
gls_results1 = gls_model1.fit()
print(gls_results1.summary())
```
